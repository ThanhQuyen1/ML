# **Task 2**. Simple Linear Regression in Scikit-Learn
---
# **Usage:**
> **Step 1**: build a model
image.png

> **Step 2**: test the model

*   With test set:
`y_pred = regressor.predict(X_test)`

*   With a single value ("predict" method always expects a 2D array as the format of its inputs)
`regressor.predict([[12]])`

> **Step 3**:
*   get w0, w1 , w2, …,
-- W0: `regressor.intercept_`
-- w1, w2, …: `regressor.coef_`
*   R^2 score: sum of squared r(i)
`r(i) = y(i) - f(x(i))`: residual of ith instance


> **Metrics**:
*   metrics.explained_variance_score(y_test, y_pred)
*   metrics.mean_absolute_error(y_ test, y_pred)
*   metrics.mean_squared_error(y_ test, y_pred[, ...])
*   metrics.mean_squared_log_error(y_ test, y_pred)
*   metrics.median_absolute_error(y_ test, y_pred)
*   metrics.r2_score(y_ test, y_pred[, ...])

# Task 2.1. dataset1.csv


```python
#code
from sklearn.linear_model import LinearRegression
regressor = LinearRegression();
dataset1 = pd.read_csv('dataset1.csv')
x1 = dataset1[['x']]
y1 = dataset1['y']
regressor.fit(x1, y1)
y_pred = regressor.predict(x1)
plt.scatter(x1, y1, color='blue')

plt.plot(x1, y_pred, color='red')
```

```python
pd.DataFrame({'y_pred': y_pred, "y" : y1})
```

```python
du_doan_don = regressor.predict([[12]])
print("Dự đoán cho giá trị [12]:", du_doan_don)
# Hệ số chặn (W0)
cham = regressor.intercept_
print("Hệ số chặn (W0):", cham)

# Các hệ số (w1, w2, ...)
he_so = regressor.coef_
print("Các hệ số (w1, w2, ...):", he_so)
from sklearn.metrics import r2_score

# Tính điểm R^2
diem_r2 = r2_score(y1, y_pred)
print("Điểm R^2:", diem_r2)
```

```python

# Điểm Phương Sai Giải Thích
print("Điểm Phương Sai Giải Thích:", metrics.explained_variance_score(y1, y_pred))

# Sai Số Trung Bình Tuyệt Đối
print("Sai Số Trung Bình Tuyệt Đối:", metrics.mean_absolute_error(y1, y_pred))

# Sai Số Bình Phương Trung Bình
print("Sai Số Bình Phương Trung Bình:", metrics.mean_squared_error(y1, y_pred))

# Sai Số Logarithm Bình Phương Trung Bình
print("Sai Số Logarithm Bình Phương Trung Bình:", metrics.mean_squared_log_error(y1, y_pred))

# Sai Số Trung Vị Tuyệt Đối
print("Sai Số Trung Vị Tuyệt Đối:", metrics.median_absolute_error(y1, y_pred))

# Điểm R^2
print("Điểm R^2 :", metrics.r2_score(y1, y_pred))
```

# Task 2.2. dataset2.csv


```python
#code
dataset2 = pd.read_csv('dataset2.csv')
from sklearn.model_selection import train_test_split

X2 = dataset2[['x1', 'x2', 'x3']]
y2 = dataset2['y']

# Chia tập dữ liệu thành tập huấn luyện và tập kiểm tra
X_train, X_test, y_train, y_test = train_test_split(X2, y2, test_size=0.3, random_state=42)

# Tạo một mô hình Hồi quy tuyến tính
regressor2 = LinearRegression()

# Khớp mô hình với tập huấn luyện
regressor2.fit(X_train, y_train)


# Dự đoán
y_pred_train = regressor2.predict(X_train)
y_pred_test = regressor2.predict(X_test)

# # Ví dụ với một giá trị đơn
# du_doan_don = regressor2.predict([[12, 15, 20]])
# print("Dự đoán cho giá trị [12, 15, 20]:", du_doan_don)

# Hệ số chặn (W0)
print("Hệ số chặn (W0):", regressor2.intercept_)

# Các hệ số
print("Các hệ số:", regressor2.coef_)

# Điểm số R^2
print("Điểm số R^2 (Tập huấn luyện):", regressor2.score(X_train, y_train))
print("Điểm số R^2 (Tập kiểm tra):", regressor2.score(X_test, y_test))
```

```python

# Điểm Phương Sai Giải Thích

print("Điểm Phương Sai Giải Thích (Tập huấn luyện):", metrics.explained_variance_score(y_train, y_pred_train))
print("Điểm Phương Sai Giải Thích (Tập kiểm tra):", metrics.explained_variance_score(y_test, y_pred_test))

# Sai số bình phương trung bình
print("Sai số bình phương trung bình (Tập huấn luyện):", metrics.mean_squared_error(y_train, y_pred_train))
print("Sai số bình phương trung bình (Tập kiểm tra):", metrics.mean_squared_error(y_test, y_pred_test))

# Sai Số Trung Vị Tuyệt Đối
print("Sai Số Trung Vị Tuyệt Đố (Tập huấn luyện):", metrics.median_absolute_error(y_train, y_pred_train))
print("Sai Số Trung Vị Tuyệt Đố (Tập kiểm tra):", metrics.median_absolute_error(y_test, y_pred_test))

# Điểm R^2 (lại)
print("Điểm R^2 (Tập huấn luyện):", metrics.r2_score(y_train, y_pred_train))
print("Điểm R^2(Tập kiểm tra):", metrics.r2_score(y_test, y_pred_test))
```

# Task 2.3. Apply LinearRegression to the following dataset:
*   Red wine quality: https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv . The dataset includes info about the chemical properties of different types of wine and how they relate to overall quality.



```python
#code
wine = pd.read_csv('winequality-red.csv')

from sklearn.model_selection import train_test_split

X3 = wine[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol']]
y3 = wine['quality']

# Chia tập dữ liệu thành tập huấn luyện và tập kiểm tra
X_train, X_test, y_train, y_test = train_test_split(X3, y3, test_size=0.3, random_state=42)

# Tạo một mô hình Hồi quy tuyến tính
regressor3 = LinearRegression()

# Khớp mô hình với tập huấn luyện
regressor3.fit(X_train, y_train)

# Dự đoán
y_pred_train = regressor3.predict(X_train)
y_pred_test = regressor3.predict(X_test)

# Hệ số chặn (W0)
print("Hệ số chặn (W0):", regressor3.intercept_)

# Các hệ số
print("Các hệ số:", regressor3.coef_)

# Điểm số R^2
print("Điểm số R^2 (Tập huấn luyện):", regressor3.score(X_train, y_train))
print("Điểm số R^2 (Tập kiểm tra):", regressor3.score(X_test, y_test))
```

```python

# Điểm Phương Sai Giải Thích

print("Điểm Phương Sai Giải Thích (Tập huấn luyện):", metrics.explained_variance_score(y_train, y_pred_train))
print("Điểm Phương Sai Giải Thích (Tập kiểm tra):", metrics.explained_variance_score(y_test, y_pred_test))

# Sai số bình phương trung bình
print("Sai số bình phương trung bình (Tập huấn luyện):", metrics.mean_squared_error(y_train, y_pred_train))
print("Sai số bình phương trung bình (Tập kiểm tra):", metrics.mean_squared_error(y_test, y_pred_test))

# Sai Số Trung Vị Tuyệt Đối
print("Sai Số Trung Vị Tuyệt Đố (Tập huấn luyện):", metrics.median_absolute_error(y_train, y_pred_train))
print("Sai Số Trung Vị Tuyệt Đố (Tập kiểm tra):", metrics.median_absolute_error(y_test, y_pred_test))

# Điểm R^2 (lại)
print("Điểm R^2 (Tập huấn luyện):", metrics.r2_score(y_train, y_pred_train))
print("Điểm R^2(Tập kiểm tra):", metrics.r2_score(y_test, y_pred_test))
```
