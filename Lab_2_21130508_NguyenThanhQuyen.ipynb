#6. Use scatter plot to represent the relationship between ***age*** and ***refletter.rating***
Rememer adding titles, xlabel, ylabel, ... to the plot

```python
# code
#ve biu 
plt.scatter(df["age"], df["refletter.rating"], marker="*")

# xlabel
plt.xlabel("age")

# ylabel
plt.ylabel("refletter")

# title
plt.title("relationship between age and refletter")

# show
plt.show()
```

#7. Use scatter plot to represent the relationship between ***age*** and ***sciences.grade***
Rememer adding titles, xlabel, ylabel, ... to the plot

```python
#code
#ve biu 
plt.scatter(df["age"], df["sciences.grade"])

# xlabel
plt.xlabel("age")

# ylabel
plt.ylabel("sciences")

# title
plt.title("relationship between age and sciences")

# show
plt.show()
```

#8. Use histogram plot to represent the distribution of ***Age***
using bins=5

```python
#code
# setting the range an bins
# range =(0, 100)
# bins = 5

# plotting a histogram
plt.hist(df["age"], bins=5, color="blue", alpha=0.8)

# xlabel
plt.xlabel("age")

# ylabel
plt.ylabel("people")

# title
plt.title("histogram plot")

#show
plt.show()
```
